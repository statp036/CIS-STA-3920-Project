```r
dataset = read.csv("student-mat.csv", header = T, na.strings="?")
```

```r
head(dataset)
```

```r
in_data= dataset[c(3,7,8,13:15,24:33)]
cor(in_data)
```

```r
summary(dataset)
```

### Best Model So Far 

```r
# 48.97 % of our predictive variables explains
# all the variation in the response variable around its mean.
best_model= lm(Walc~failures+Dalc+absences+goout,data=dataset)
summary(best_model)
```

```r
plot(best_model)
```

```r
plot(fitted(best_model), residuals(best_model))

#add horizontal line at 0
abline(h = 0, lty = 2)
```

```r
# We can see that the best model almost follows a normal distribution.
hist(residuals(best_model), col = "steelblue")
```
